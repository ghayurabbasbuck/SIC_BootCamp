```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv('elections.csv')
```

```python
df.head()
```

### Using loc

##### 1.  Extract all rows for the candidate 'Andrew Jackson' and display the columns 'Year', 'Popular vote', and 'Result'

```python
df.loc[df['Candidate']=='Andrew Jackson',['Year','Popular vote','Result']]
```

##### 2. Use `loc` to extract all rows where the candidate's `Party` is 'Democratic' and display the 'Year', 'Candidate', and 'Result'.

```python
df.loc[df['Party'] == 'Democratic',['Year','Candidate','Result']]
```

##### 3. Extract rows where the `Popular vote` is greater than 600000 and display the 'Candidate' and 'Party' columns.

```python
df.loc[df['Popular vote'] > 600000,['Candidate','Party']]
```

##### 4. Use `loc` to find all candidates who won the election and display their 'Year' and 'Candidate'.

```python
df.loc[df['Result']=='win', ['Year','Candidate']]
```

##### 5. Extract the 'Year', 'Candidate', and 'Percentage' for candidates from the 'National Republican' party using `loc`.

```python
df.loc[df['Party'] == 'National Republican', ['Year','Candidate','%']]
```

### Using iloc

##### 6. Retrieve the first 5 rows and display only the first 3 columns using `iloc`.

```python
df.iloc[:5,:3]
```

##### 7. Use `iloc` to extract the 3rd to 7th rows and display all columns

```python
df.iloc[3:7,:]
```

##### 8. Use `iloc` to retrieve the last 5 rows and display only the 'Year' and 'Candidate' columns.

```python
df.iloc[-5:,[0,1]]
```

##### 9. Extract rows from index 2 to 6 and display the 'Popular vote' and 'Percentage' columns using `iloc`.

```python
df.iloc[2:6,[3,5]]
```

##### 10. Use `iloc` to display the first 10 rows but only show the 'Party' and 'Result' columns.

```python
df.iloc[:10,[2,4]]
```

### Using []

##### 11. Display only the `Party` and `Result` columns for candidates whose `Popular vote` is greater than 500,000.

```python
df[df['Popular vote'] > 500000][['Party','Result']]
```

##### 12. Extract all rows where the `Year` is greater than 1900 and display the 'Candidate' and 'Popular vote' columns.

```python
df[df['Year']>1900][['Candidate','Popular vote']]
```

##### 13. Extract the 'Year' and 'Percentage' for candidates whose percentage is greater than 50.

```python
df[df['%']>50][['Year','%']]
```

##### 14. Display only the rows where the 'Result' is 'loss' and show the 'Candidate' and 'Popular vote' columns.

```python
df[df['Result']=='loss'][['Candidate','Popular vote']]
```

##### 15. Extract rows where the 'Party' is 'Democratic-Republican' and display all columns.

```python
df[df['Party']=='Democratic-Republican'][:]
```
